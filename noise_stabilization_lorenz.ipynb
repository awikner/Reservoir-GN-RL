```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_times = [50,100,200,300,400,500]
res_size = 100
noise_values = np.linspace(-3.666666666666, 0, num = 12)

fig = plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 16})
for train_time in train_times:
    stable_frac = np.loadtxt('stable_frac_%dnodes_%dtrainiters.csv' %(res_size, train_time), delimiter = ',')
    plt.plot(noise_values, stable_frac, 'x-', label = 'T = %d' % train_time)
plt.xlabel(r'$\log_{10}$(Noise)')
plt.ylabel('Fraction of Stable Predictions')
plt.title('Stable Predictions for N=100 Reservoir')
plt.legend()
plt.savefig('stable_fraction.png')
plt.show()
```

```python
train_times = [50,100,200,300,400,500]
res_size = 100
noise_values = np.linspace(-3.666666666666, 0, num = 12)

fig = plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 16})
for train_time in train_times:
    stable_frac = np.loadtxt('stable_frac_alpha_%dnodes_%dtrainiters.csv' %(res_size, train_time), delimiter = ',')
    plt.plot(noise_values, stable_frac, 'x-', label = 'T = %d' % train_time)
plt.xlabel(r'$\log_{10}$(Noise)')
plt.ylabel('Optimal Regularization for Stability')
plt.yscale('log')
plt.title('Stable Predictions for N=100 Reservoir')
plt.legend()
plt.savefig('best_stable_regularization.png')
plt.show()
```
